# Machine Learning

```python
from sklearn.cluster import KMeans
```

```python
def calcular_wcss(dados_cliente):
  wcss= []
  for k in range(1,11):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(X=dados_cliente)
    wcss.append(kmeans.inertia_)
  return wcss
```

```python
#Separando as variáveis que serão levadas em conta para clusterização
#dados_cliente = base_completa[[ 'sexo', 'idade', 'qtd_filhos', 'estado_civil', 'cod_estado', 'hipertenso', 'diabetes', 'cod_produto', 'cod_classe_produto']]
#dados_cliente = base_completa[[ 'sexo', 'idade', 'hipertenso', 'diabetes']]
dados_cliente = base_completa[[ 'idade']]

dados_cliente.head()
```

# Analisando WCSS
##Métrica utilizada na análise de agrupamento (clustering) para avaliar a qualidade dos clusters que seão necesários



```python
wcss_clientes = calcular_wcss(dados_cliente)
```

```python
type(wcss_clientes)
```

```python
for i in range(len(wcss_clientes)):
  print(f'O Cluester{i} possui valor de wcss de : {wcss_clientes[i]}')
```

```python
import plotly.express as px
import plotly.offline as py
import plotly.graph_objects as go
```

```python
grafico_wcss = px.line( x= range(1,11),
                        y= wcss_clientes)

fig = go.Figure(grafico_wcss)

fig.update_layout(title='Calculando o WCSS',
                 xaxis_title= 'Número de clusters',
                 yaxis_title= 'Valor do Wcss',
                 template = 'plotly_white'
                  )
fig.show()
```

```python
#Percebemos pelo gráfico que a curva fica linear após o cluster 3, portanto esse é o numero de cluster que utilizaremos para separar
```

```python
#Treinando o modelo
kmeans_clientes = KMeans(n_clusters=3, random_state=0)
#Inserindo os cluster a base_completa
base_completa['cluster'] = kmeans_clientes.fit_predict(dados_cliente)
```
